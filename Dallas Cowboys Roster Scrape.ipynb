```python
import requests
from bs4 import BeautifulSoup

# Specify the URL you want to scrape
url = "https://www.nfl.com/teams/dallas-cowboys/roster"  # Replace this with the actual URL

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text
    print("Page content fetched successfully!")
    
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Example: Find all hyperlinks on the page
    links = soup.find_all("a")
    for link in links:
        print(link.get("href"))
else:
    print("Failed to retrieve the page")
```

```python
import requests
from bs4 import BeautifulSoup

# URL for the Dallas Cowboys roster (replace with the correct page URL)
url = "https://www.nfl.com/teams/dallas-cowboys/roster"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the table or section containing player info
    players = soup.find_all("tr")  # Adjust based on the HTML structure

    # Extract desired information
    for player in players:
        # Assuming each player's data is in <td> tags in the correct order
        columns = player.find_all("td")
        
        if columns:  # Ensure there are <td> elements to prevent errors
            # Extracting data based on position in the row
            name = columns[0].text.strip()  # Adjust index if needed
            jersey_number = columns[1].text.strip()  # Adjust index if needed
            position = columns[2].text.strip()
            experience = columns[-2].text.strip()
            college = columns[-1].text.strip()  # Assuming college is in the last column
            
            print(f"Name: {name}, Jersey Number: {jersey_number}, Position: {position} ,Experience in Years: {experience} , College: {college}")
else:
    print("Failed to retrieve the page") 
```
